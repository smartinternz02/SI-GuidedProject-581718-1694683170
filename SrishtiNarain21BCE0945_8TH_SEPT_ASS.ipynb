```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv("car_crashes.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

LINE GRAPH


```python
x=dataset['speeding']
y=dataset['total']
plt.plot(x,y,'b-*')
```

Inference:from the plot we can say that as speeding increases total is also increasing gradually

SUBPLOTS

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b--')
plt.subplot(2,2,2)
plt.plot(x,y,'r*')
plt.subplot(2,2,3)
plt.plot(x,y,'g*')
plt.subplot(2,2,4)
plt.plot(x,y,'r>')
```

Inference: same data representation in subplot

SCATTER PLOTS

```python
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
```

Inference: same data representation in scatterplot

BAR CHART

```python
plt.bar(x,y,width=0.1)
```

```python
plt.barh(x,y,color="red")
```

Inference: same data representation in bar graph

HISTOGRAM

```python
plt.hist(y)
plt.show()
```

Inference: total value mostly occurs between 17.5 and 20

PIE CHART

```python
dataset["total"].value_counts()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.pie(dataset["total"],autopct="%0.2f%%")
axes1.legend()
```

Inference: total attribute shown in the form of pie chart---all values occur almost equally

```python
sns.scatterplot(x="speeding",y="total",data=dataset)
```

Inference:from the plot we can say that as speeding increases total is also increasing

```python
sns.lineplot(x="speeding",y="total",data=dataset,ci=None)
```

Inference: same data representation in line plot

```python
sns.distplot(dataset["speeding"])
```

Inference: speed value having max density is between 4 & 5

```python
sns.relplot(x="total",y="speeding",data=dataset,hue="abbrev")
```

Inference: scatterplot between total and speeding attribute on the basis abbreviations they are divided into

```python
sns.barplot(data=dataset,x="abbrev",y="speeding",ci=None)
```

Inference: car with abbrev HI has max speed

```python
sns.countplot(x="alcohol",data=dataset)
```

Inference: there were only 2 people who had same amount of alcohol content during crash

```python
sns.jointplot(x="speeding",y="total",data=dataset)
```

Inference: same data representation in joint grid

```python
sns.boxplot(x="abbrev",y="total",data=dataset)
```

Inference: abbrev presentation using boxplot

```python
corr=dataset.corr()
corr
```

Inference: correlation amongst all the attributes

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```
